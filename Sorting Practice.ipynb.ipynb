# Selection Sort

```python
def selection_sort(arr,n):
    for i in range(0,n-1):
        # select the min index
        min = i
        # check which index is the minimum
        for j in range(i,n):
            if arr[j] < arr[min]:
                min = j
        
        # swap the i with min
        arr[i],arr[min] = arr[min],arr[i]
    return arr
    
```

# Bubble Sort

```python
def bubble_sort(arr,n):
    for i in range(n-1,0,-1):
        for j in range(0,i):
            if arr[j] > arr[j+1]:
                arr[j],arr[j+1] = arr[j+1],arr[j]
                
    return arr
```

# Insertion Sort

```python
def insertion_sort(arr,n):
    for i in range(0,n):
        j = i
        while (j > 0 and arr[j-1] > arr[j]):
            arr[j-1],arr[j] = arr[j],arr[j-1]
            j = j-1
        
    return arr
        
        
```

```python
arr = [9,12,6,7,14,3]
n = len(arr)

sorted_arr = insertion_sort(arr,n)
print(sorted_arr)
```
